```python
# Implementacion de la clase Animal

class Animal:
  def __init__(self, edad = 0, nombre = "Godzilla"):
    """
    Constructor de la clase Animal

    :param edad : Edad del animal : Por defecto tiene el valor de cero
    :param nombre : Nombre del animal : Por defecto tiene el valor de Godzilla
    :return : UNa instancia de la clase Animal
    """

    # Todos los atributos de este constructor sean publicos
    self.edad = edad
    self.nombre = nombre
    self.sexo = "F"
    self.vivo = True
    # Animal.numAnimals = Animal.numAnimals +1

  def saluda():
    print("Hola Mundo")

  def showmethename(self):
    print("Bienvenido " + self.nombre)

  def showmetheage(self):
    print("Edad " + str(self.edad))
```
